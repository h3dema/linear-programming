# Programacao linear

A programação linear é um conjunto de técnicas usadas para resolver sistemas de equações lineares e desigualdades enquanto maximiza ou minimiza alguma função linear. É importante em áreas como computação científica, economia, ciências técnicas, manufatura, transporte, militar, gerenciamento, energia e assim por diante.


### Exemplo de sistema de equacoes lineares

Sistemas lineares são formados por duas ou mais equações lineares que possuem suas incógnitas relacionadas.
O trabalho com equações existe devido à necessidade de encontrarmos valores desconhecidos de incógnitas. Chamamos de equação quando temos uma expressão algébrica com igualdade, e ela é classificada como linear quando o maior expoente de suas incógnitas é 1.
De modo geral, uma equação linear pode ser descrita por:

$a_1 x_1 + a_2 x_2 + a_3 x_3 \dots + a_n x_n = c$

Um sistema de equacoes contem diversas equacoes lineares como no exemplo abaixo:

$\begin{matrix}
2x_1 + x_2 = 3 \\
-x_1 + 2x_2 > 0 \\
3x_1 + 4x_2 > 5 \\
x_1 > 0 \\
x_2 > 0
\end{matrix}$


### O que é Programação Linear?

Imagine que você tem um sistema de equações lineares e inequações. Tais sistemas geralmente têm muitas soluções possíveis. A programação linear é um conjunto de ferramentas matemáticas e computacionais que permite encontrar uma solução particular para esse sistema que corresponda ao máximo ou mínimo de alguma outra função linear.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

f1 = lambda x: 3 - 2 * x
f2 = lambda x: x / 2
f3 = lambda x: (5 - 3 * x) / 4

# Eq.2
plt.fill([0, 100, 100, 0],[f2(0), f2(100), 100, 100], 'red', alpha=0.2)
plt.text(2, 3, "$-x_1 + 2x_2 > 0$", fontsize=16)


# Eq.3
plt.fill([0, 100, 100, 0],[f3(0), f3(100), 100, 100], 'blue', alpha=0.2)
plt.text(2, 0.5, "$3x_1 + 4x_2 > 5$", fontsize=16)


# Eq.1
plt.plot([0, 100], [f1(0), f1(100)], color="black")


plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$x_2$", fontsize=16)
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.show()
```
